```python
import matplotlib.pyplot as plt

# Example data: (time in hours, code coverage percentage)
data = [(1, 54), (2, 55), (3, 55), (4, 65), (5, 67)]

x, y = zip(*data)

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.xlabel('Time (hours)')
plt.ylabel('Code Coverage (%)')
plt.title('Code coverage of WolfBoot')
plt.grid(True)
plt.show()
```
